## Section 2: Exploratory Data Analysis

How can descriptive Statistics Give You a Broad Overview of Data? What Descriptive Statistics Would You Use?

It can allow you to summarize the spreads, deviations, ranges, averages, and visual representations of the data. Specifically we can use, mean, median, mode, range, standard deviation, outliers, etc. Particular to my dataset, using mean, median, and standard deviation is ideal as the range and outliers and very limited.

```python
len(df)
```

```python
data = []
d = {"Player-additional": "Mean","PTS" :df["PTS"].mean(), "FGA" :df["FGA"].mean(), "FG":df["FG"].mean(), "TRB":df["TRB"].mean(),"AST":df["AST"].mean(), "TOV":df["TOV"].mean()}
d1 = {"Player-additional": "Median","PTS" :df["PTS"].median(), "FGA" :df["FGA"].median(), "FG":df["FG"].median(), "TRB":df["TRB"].median(),"AST":df["AST"].median(), "TOV":df["TOV"].median()}
d2 = {"Player-additional": "Standard Deviation","PTS" :df["PTS"].std(), "FGA" :df["FGA"].std(), "FG":df["FG"].std(), "TRB":df["TRB"].std(),"AST":df["AST"].std(), "TOV":df["TOV"].std()}

data.append(d)
data.append(d1)
data.append(d2)
Analysis = pd.DataFrame(data)
Analysis
```

All of the means are greater than the medians. This implies that there is a greater frequency of higher stats. Additionally the standard deviation is relatively low for most statistics implying that the data is not deviating too much. ## Section 3: Exploratory Data Visualization

```python

#Targetting categories to create more quality plots
df_target_categories = df[["Player", "Pos", "PTS", "AST", "TRB", "MP", "FG%","TOV"]]
df_target_categories["STL + BLK"] = df["STL"] + df["BLK"]
df_target_categories
```

```python
df_group = df_target_categories.groupby("Pos").agg("mean", numeric_only = True).reset_index()
df_group
```

```python

df_group.plot.line(x="Pos", y ="PTS", xlabel = "Position", ylabel = "Points Averaged", title = "Points Scored based on Position")
```

```python
df_group.plot.line(x="Pos", y ="AST", xlabel = "Position", ylabel = "Assists Averaged", title = "Assists Scored based on Position")
```

```python
df_group.plot.bar(x="Pos", y ="TRB", xlabel = "Position", ylabel = "Rebounds Averaged", title = "Rebounds Scored based on Position")
```

```python
df_group.plot.bar(x="Pos", y ="MP", xlabel = "Position", ylabel = "Minutes Averaged", title = "Minutes Played based on Position")
```

## Section 5: Data Science

The question I have is: How can we predict the number of assists a point guard gets based on the rebounds they average?

I tweaked this question to explore point guards specifically as they had good correlation between stats during prior analysis.

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
df_pg = df_more_5_min_played[df_more_5_min_played.Pos == "PG"]
ind = ["TRB"]
dep = "AST"
model.fit(df_pg[ind],df_pg[dep]) 
model.predict(df_pg[ind])
```

```python
df_pg["TRB"].corr(df_pg["AST"])
```

## Section 6: A Different Data Visualization

```python
import seaborn as sns # I got this from lab regression
import matplotlib.pyplot as plt

sns.regplot(data=df_pg, x="TRB", y="AST", ci=None, line_kws={'color':'red'})
plt.title("Total Rebounds predicting Assists of a point guard")
plt.xlabel("Rebounds")
plt.ylabel("Assists")
```

This visualization shows a linear regression model predicting the assists based on rebounds by point guards. This Linear Regression model shows a strong positive correlation between rebounds and assists. This shows that active point guards who are involved with rebounding couples well with the amount of assists they get. I also notice that point guards average more assists than rebounds as the slope is evidently greater than 1.
